#### 1. Carregando os conjuntos de dados "gasolina_2000+.csv" e "gasolina_2010+.csv" em dois DataFrames diferentes e combinando em um único DF

```python
import pandas as pd
```

#### 4.Convertendo Colunas "DATA INICIAL E DATA FINAL" Para data

#### 5. Criando uma nova coluna para representar o ano e o mês(aaaa-mm), utilizando a coluna DATA FINAL como referência.

```python
df["ANO-MES"] = df["DATA FINAL"].apply(lambda x: '{}'.format(x.year)) + df["DATA FINAL"].apply(lambda x: '-{:02d}'.format(x.month))
```

#### 8. Preço médio de revenda da gasolina em agosto de 2008

```python
df2[df2['ANO-MES'] == '2008-08']["PREÇO MÉDIO REVENDA"].mean()
```

#### 9. Qual o preço médio da gasolina em maio de 2014 em São Paulo

```python
df2[(df2['ANO-MES'] == '2014-05') & (df2['ESTADO'] == 'SAO PAULO')]["PREÇO MÉDIO REVENDA"].mean()
```

#### 10.Descobrindo em qual estado a gasolina passou dos R$ 5,00 E Quando isso aconteceu

#### 11. Qual a Média de Preço dos estados da região sul em 2012?

```python
df_aux = df2[df2["DATA FINAL"].apply(lambda x: x.year) == 2012]
df_aux[df_aux['REGIÃO'] == 'SUL']["PREÇO MÉDIO REVENDA"].mean()
```

#### 12. Tabela com a variação percentual ano a ano para o Rio de Janeiro

```python
df2["MES"] = df2["DATA FINAL"].apply(lambda x: x.month)
df_rio = df2[df2["ESTADO"] == "RIO DE JANEIRO"]
```

```python
df_month_rio = df_rio.groupby("ANO-MES")[["PREÇO MÉDIO REVENDA", "MES"]].last()
```

```python
(df_month_rio[df_month_rio["MES"] == 12] / df_month_rio[df_month_rio["MES"] == 12].shift(1) - 1) * 100
```
